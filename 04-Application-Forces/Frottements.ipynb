## Mouvement dans un fluide, avec frottements

$v(t)=v_0e^{-\lambda t}$


```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
```

```python
vzero=10
lamb=.25
t = np.linspace(0,20,1000)
vdet=vzero*np.exp(-lamb*t)
tangente=vzero-lamb*vzero*t
matplotlib.rcParams['figure.figsize'] = (8,4)
plt.plot(t,vdet,lw=3)
plt.plot(t,tangente)
plt.grid(True)
plt.xlabel('temps',fontsize=12)
plt.ylabel('v(t)',fontsize=12)
plt.tick_params(labelleft=False,labelbottom=False)
plt.annotate('$v_0$',xy=(-2,9.8),annotation_clip=False,fontsize=12)
plt.ylim(bottom=-1)
plt.ylim(top=11)
plt.show()
```

```python
g=10
v2det=g/lamb*(1-np.exp(-lamb*t))
```

```python
plt.plot(t,v2det,lw=3)
plt.grid(True)
plt.xlabel('temps (s)',fontsize=12)
plt.ylabel('v(t) (m/s)',fontsize=12)
plt.show()
```
