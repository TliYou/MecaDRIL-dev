## Orbites des satellites et autres

```python
%matplotlib notebook
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets

G=6.67*10**(-11) #cte gravitation m^3kg^-1s^-2
MT=5.97*10**24 # masse terre kg
RT=6.4*10**6 # Rayon Terre m
```

```python
def vorb(h):
    vo=np.sqrt(2*G*MT/(RT+h*1000))
    return vo

def torb(h):
    to=2*np.pi*np.sqrt((RT+h*1000)**3/(G*MT))/3600
    return to
```

```python
torb(0)
```

```python
h = np.linspace(0,380000,100000)
plt.plot(h,torb(h))
plt.grid(True)
plt.xlabel('altitude (km)',fontsize=12)
plt.ylabel('pÃ©riode (heures)',fontsize=12)
plt.show()
```
